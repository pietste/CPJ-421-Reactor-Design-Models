```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
from lmfit import Minimizer, Parameters, report_fit
%matplotlib notebook
```

```python
M = np.matrix(pandas.read_excel('Dataminus.xlsx'))
time = M[:,0]/60
lim1 = M[:,1]
def rate_release(k1, k2):
    M_t = k1*t**0.5 + k2*t
    return M_t

t = np.linspace(0,1000,10000)
Z = (time,lim1)
def func(params, Z):
    Z = time,lim1
    k1 = params['k1']
    k2 = params['k2']
    def rate_release(k1, k2):
        M_t = k1*t**0.5 + k2*t
        return M_t
    M_t = rate_release(k1,k2)
    Mlim1 = np.interp(time,t,M_t)
    ERROR = abs(lim1 - Mlim1)
    return ERROR
def Model1(k1,k2,Z):
    Z = time,lim1
    def rate_release(k1, k2):
        M_t = k1*t**0.5 + k2*t
        return M_t
    M_t = rate_release(k1,k2)
    plt.figure('d-limonene sample 1')
    plt.plot(time,lim1, marker='o', color='darkgreen', linestyle='', label='Experimental data')
    plt.plot(t, M_t, color='darkgreen', label='Fitted model')
    plt.xlabel('Time (h)')
    plt.ylabel('Mass fraction released')
    plt.axis([0,100,0,1])
    plt.legend(loc='best')
    plt.show()
    plt.figure('Compare')
    plt.plot(time,lim1, marker='o',markeredgecolor='none', color='green', linestyle='', label='Sample 1 data')
    plt.plot(t, M_t, color='green', label='Model 2.1')
    plt.xlabel('Time (h)')
    plt.ylabel('Mass fraction released ($M(t)/M_\infty$)')
    plt.axis([0,100,0,1])
    plt.legend(loc='best')
    plt.show()

params = Parameters()
params.add('k1',   value= 0.00001,  min=0, max= 5)
params.add('k2',   value= 0.00001,  min=0, max=5)
minner = Minimizer(func, params,fcn_args=(Z,))
result1 = minner.minimize()
k1 = result1.params['k1']
k2 = result1.params['k2']
print(report_fit(result1))
Model1(k1,k2,Z)    
```
